### Importing in libs

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

### Load the webpage

```python
r = requests.get('https://gobobcats.com/sports/baseball/stats?path=baseball')

#Convert to a beautsoup object
soup = bs(r.content)

contents = soup.prettify()
print(contents)
```

### Getting the correct headers from the webpage

```python
headers = [th.getText() for th in soup.findAll('tr')[0].findAll('th')] #Find the second table row tag, find every table header column within it and extract the html text via the get_text method.
headers = headers[1:] #Do not need the first (0 index) column header
print(headers[:])
```

### Getting the rows that contain player stats

```python
rows = soup.findAll('tr') #Here we grab all rows that are not classed as table header rows 
player_stats = [[td.getText() for td in rows[i].findAll('td')] # get the table data cell text from each table data cell
                for i in range(len(rows))] #for each row
player_stats = player_stats[1:18] #selecting the correct ammount of rows 
print(player_stats)
```

```python
stats = pd.DataFrame(player_stats, columns = headers) #creating a df for the scraped data
stats
```

```python
stats.to_csv('2022playerstats.csv') #creating a cvs for the player stats
```
